```python
import pandas as pd
```

```python
df1= pd.read_csv('df_2016_2018.csv')
df1= df1[['Borough','Year','Crime rate', 'Population per hectare', 'Consumer expenditure on Pubs and Wine bars (Â£mn)','Health', 'Public Admin and defence','Dwellings per hectare']]
df1= df1.rename(columns={'Health': '% of Health workers', 'Public Admin and defence': '% of Public Admin and defence workers'})

#group df1 by 'Borough' and computes mean of each column for each borough and assign the resultant dataframe to df1_avg
df1_avg = df1.groupby('Borough').mean()

#Rename elements in the Year column, and rename the Year column to Timeframe
df1_avg.Year= ['Mean values for 2016-2018' for i in range(32)]
df1_avg= df1_avg.rename(columns={ 'Year':'Timeframe'})
df1_avg.iloc[:,1:] = df1_avg.iloc[:,1:].round(decimals=2)

df1_avg
```

```python
#Save csv
df1_avg.to_csv('mean_top_five_important_attributes_2016_2018.csv', index=True)
```
