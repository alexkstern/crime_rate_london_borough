```python
import pandas as pd
```

```python
df1= pd.read_csv('crime_rate_2016_2021.csv')
df1
```

```python
df2= pd.read_csv('crime_rate_1999_2016.csv')
df2= df2.drop('2016', axis=1)
df2
```

```python
#Merge the two data frames
df = pd.merge(df2, df1, on='Borough')
df= df.drop('Code',axis=1)
# Calculate the mean of the columns from index 1 to the end
mean_column = df.iloc[:, 1:].mean(axis=1)

# Add the mean column to the DataFrame
df['mean_crime_rate'] = mean_column

# Display the DataFrame
df
```

```python
df.to_csv('crime_rate_1999_2021.csv')
```
